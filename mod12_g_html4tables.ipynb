  <h1 class="page-title">Commercial Systems in IO</h1>
  
<p class="mt-md">Why focus on commercial systems in a course on the organization of information?&nbsp;</p>
<ul>
<li>The move to online resources and the need for systems to organize them has led to the adoption of numerous software products that provide different avenues for access.</li>
<li>Systems may be created in house, may use open source software, or may be commercial systems. There are pros and cons to each one.</li>
<li>Some commercial products are offered by huge conglomerates, with broad commercial interests. This poses potential issues for libraries. Issues related to IO activities have been selected for coverage in the module</li>
</ul>
<h2 class="mt-md" style="color: #de5b49;"><strong>Identification of Problems</strong></h2>
<p>IO concerns with commercial products include privacy of user information, the monetization of information, and bias in search results.</p>
<p>In a 2020 report on information discovery and use, Sarah Lippincott (2020) highlights some of these issues:</p>
<blockquote class="ml-md mr-md mt-lg mb-lg p-all-md pl-xl" style="border-left: 2px dashed #bdbcbc;">
<p style="line-height: 1.8;">ML [machine learning] and NLP [natural language processing] technologies will facilitate new forms of search, discovery, and academic inquiry. At best, these technologies create exciting new modes of inquiry, facilitate cross-disciplinary discovery, and make research more efficient and productive. However, they have the potential to suppress human agency in the research process, amplify biases, and expose users to data-privacy violations. (p. 25)</p>
</blockquote>
<p>She also comments that while recommender systems can assist in the research process, "they rely on intensive collection and analysis of user data, which can compromise user privacy in ways that are anathema to most libraries" (p. 11).</p>
<p>As an example of early concerns about commercial entities, when Google partnered with large research libraries to digitize their collections, Rory Litwin (2007) noted the following issues in an opinion piece: The monetization of library collections, privacy concerns, and the "bias introduced by commercialization<span>" (section, The Bias). </span><span style="font-family: inherit; font-size: 1rem;">&nbsp;</span></p>
<h2 class="mt-md" style="color: #de5b49;"><strong>Adoption of Commercial Systems</strong></h2>
<p>The automation of many library services has developed a market for a variety of commercial software products, including systems that provide discovery and access, and the storage and preservation of digital files. While open source products are an alternative, they are not always feasible for a library to adopt as they may require inhouse programming and expertise to implement and manage. In between open-source and products offered by for-profit commercial first are companies like OCLC. OCLC is nonprofit, membership organization. Subscription and membership fees are sometimes not within reach of smaller libraries.&nbsp;</p>
<p>Commercial products may be adopted because they are the only alternative or are best alternative.&nbsp;</p>
