# Overview of Archives and Museums, from an IO Lens

The last Module looked at some big-picture ideas related to retrieval, and this Module has focused so far on library catalogs.

Naturally, the kinds of retrieval systems in use will depend on the context—_What kind of information agency? What kind of content? What expectations do users have?_ Presumably, the approaches in other information agencies will differ from the approaches used in libraries for these reasons.

The rest of this Module branches out to consider organization and retrieval in different kinds of information agencies. Many information agencies have libraries—special libraries exist in museums, and often are part of archives as well. The focus this Module so far has been on libraries, or the organization of written physical materials for access by others. To get us started thinking beyond libraries, this brief introduction will look at both archives and museums as information agencies/cultural heritage institutions in which organized information is made available to users.

## Archives and Archival Studies, the Basics

The past few weeks have focused on theory and practice that has largely evolved into work carried out in libraries and other information agencies. But what about archives and museums in particular? How are they similar? How are they different?

#### Archives

Information professionals work in a variety of settings—anywhere there's "information"! This includes _archives_. The ODLIS states the following about [archives](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archives):

> An organized [collection](https://www.abc-clio.com/ODLIS/odlis_c.aspx#collection) of the [noncurrent](https://www.abc-clio.com/ODLIS/odlis_n.aspx#noncurrent) [records](https://www.abc-clio.com/ODLIS/odlis_r.aspx#records) of the activities of a business, government, organization, institution, or other [corporate body](https://www.abc-clio.com/ODLIS/odlis_c.aspx#corporatebody), or the [personal papers](https://www.abc-clio.com/ODLIS/odlis_p.aspx#personalpapers) of one or more individuals, families, or groups, [retained](https://www.abc-clio.com/ODLIS/odlis_r.aspx#retention) permanently (or for a designated or indeterminate period of time) by their originator or a successor for their permanent historical, [information](https://www.abc-clio.com/ODLIS/odlis_i.aspx#information)al, evidential, legal, administrative, or monetary value, usually in a [repository](https://www.abc-clio.com/ODLIS/odlis_r.aspx#repository) managed and maintained by a trained [archivist](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivist) (see [this example](http://graphics8.nytimes.com/images/2007/05/11/travel/venues-lsociety-library.jpg)). Also refers to the office or organization responsible for [appraising](https://www.abc-clio.com/ODLIS/odlis_a.aspx#appraisal), selecting, [preserving](https://www.abc-clio.com/ODLIS/odlis_p.aspx#preservation), and providing [access](https://www.abc-clio.com/ODLIS/odlis_a.aspx#access) to archival materials.
> 
> Archives can be classified in three broad categories: [government archives](https://www.abc-clio.com/ODLIS/odlis_g.aspx#govarchives) (_example_: [National Archives and Records Administration](https://www.abc-clio.com/ODLIS/odlis_n.aspx#nara)), [in-house archives](https://www.abc-clio.com/ODLIS/odlis_i.aspx#inhousearchives) maintained by a parent institution, and [collecting archives](https://www.abc-clio.com/ODLIS/odlis_c.aspx#collectingarchives) ([manuscript](https://www.abc-clio.com/ODLIS/odlis_m.aspx#manuscript) [libraries](https://www.abc-clio.com/ODLIS/odlis_l.aspx#library), [film archives](https://www.abc-clio.com/ODLIS/odlis_f.aspx#filmarchives), [genealogical](https://www.abc-clio.com/ODLIS/odlis_g.aspx#genealogy) archives, [sound archives](https://www.abc-clio.com/ODLIS/odlis_s.aspx#soundarchives), [personal archives](https://www.abc-clio.com/ODLIS/odlis_p.aspx#personalarchives), etc.). [ProQuest](https://www.abc-clio.com/ODLIS/odlis_p.aspx#proquest) provides the [subscription](https://www.abc-clio.com/ODLIS/odlis_s.aspx#subscription) [database](https://www.abc-clio.com/ODLIS/odlis_d.aspx#database) [_Archive Finder_](http://www.proquest.com/en-US/catalogs/databases/detail/archives_usa.shtml). Compare with [archive](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archive). **_See also_**: [archival copy](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivalcopy), [archival database](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivaldatabase), [archival jurisdiction](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivaljuris), [archival paper](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivalpaper), [archival quality](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivalquality), [archival value](https://www.abc-clio.com/ODLIS/odlis_a.aspx#archivalvalue), [artificial collection](https://www.abc-clio.com/ODLIS/odlis_a.aspx#artificialcollec), [digital archives](https://www.abc-clio.com/ODLIS/odlis_d.aspx#digitalarchives), [International Council on Archives](https://www.abc-clio.com/ODLIS/odlis_i.aspx#ica), and [Society of American Archivists](https://www.abc-clio.com/ODLIS/odlis_s.aspx#saa).
> 
> The [term](https://www.abc-clio.com/ODLIS/odlis_t.aspx#term) is also used in academia to refer to a [repository](https://www.abc-clio.com/ODLIS/odlis_r.aspx#repository) of electronic [preprints](https://www.abc-clio.com/ODLIS/odlis_p.aspx#preprint), [working paper](https://www.abc-clio.com/ODLIS/odlis_w.aspx#workingpaper)s, and similar [document](https://www.abc-clio.com/ODLIS/odlis_d.aspx#document)s, commonly called _[e-print](https://www.abc-clio.com/ODLIS/odlis_e.aspx#eprint) archives_. Used in this sense, there is no implication of archival management, which has caused some confusion, for example, around the purpose of the [Open Archives Initiative](https://www.abc-clio.com/ODLIS/odlis_o.aspx#oai) (OAI).

The term _archives_ can clearly mean a lot of things. Of those primary three categories, some examples are worth exploring.

*   **[government archives](https://www.abc-clio.com/ODLIS/odlis_g.aspx#govarchives)**—Since the definition already pointed to the US national archives, another governmental archives would be a good point of comparison: Australia's [https://www.naa.gov.au/](https://www.naa.gov.au/)
*   **[in-house archives](https://www.abc-clio.com/ODLIS/odlis_i.aspx#inhousearchives) (maintained by a parent institution)**—truly, any _house_ or parent institution can have its own archives. Universities do. For-profit companies do. Even the Smithsonian does. Take a look around the website for the Smithsonian: [https://siarchives.si.edu/](https://siarchives.si.edu/)
*   [**collecting archives**](https://www.abc-clio.com/ODLIS/odlis_c.aspx#collectingarchives)—Under the umbrella of their [archival collections](https://library.umkc.edu/archival-collections), the University of Missouri Kansas City houses the Marr Sound Archive: [https://library.umkc.edu/marr](https://library.umkc.edu/marr)

#### Archival Science

As with the work done in any information environment, there is a body of theory that has evolved through work in archives. The Society of American Archivists defines [_archival science_ ](https://www2.archivists.org/glossary/terms/a/archival-science)in the following way, and provides the term in a citation from Duranti and MacNeil:

> 1.  ~ A systematic body of theory that supports the practice of appraising, acquiring, authenticating, preserving, and providing access to recorded materials.
> 
> **Citations:** †(Duranti & MacNeil, 1996, p. 47) Archival science, which emerged out of diplomatics in the nineteenth century, is a body of concepts and methods directed toward the study of records in terms of their documentary and functional relationships and the ways in which they are controlled and communicated.

All of the activities carried out in archives are included in archival studies, but naturally, in terms of this class, the primary concern is with the mechanisms for organizing archival materials and the retrieval tools that make those materials findable, retrievable, and usable.

**Museums and Museum Studies, the Basics**
------------------------------------------

Despite the differences in the purpose and function of museums and archives, the two are nonetheless very tightly tied. Often, archives can grow into museums. Have you followed the origins of the new Academy Museum of Motion Pictures ([https://www.academymuseum.org/en/)](https://www.academymuseum.org/en/) in Los Angeles? It grew out of the archival collections of the Motion Picture Association ([https://www.motionpictures.org/](https://www.motionpictures.org/)).

#### Museums

So how exactly are museums different from archives? Keeping in mind what was presented about museums, it is worth now considering museums. The ODLIS defines [museums](https://www.abc-clio.com/ODLIS/odlis_m.aspx#museum) this way:

> A publicly or privately funded nonprofit institution whose primary function is the [preservation](https://www.abc-clio.com/ODLIS/odlis_p.aspx#preservation) and display of [collection](https://www.abc-clio.com/ODLIS/odlis_c.aspx#collection)s of physical [artifact](https://www.abc-clio.com/ODLIS/odlis_a.aspx#artifact)s and [specimen](https://www.abc-clio.com/ODLIS/odlis_s.aspx#specimen)s for the purposes of education, scholarship, and enjoyment. Since [book](https://www.abc-clio.com/ODLIS/odlis_b.aspx#book)s and [binding](https://www.abc-clio.com/ODLIS/odlis_b.aspx#binding)s are physical artifacts, some museums include them in their collections, for example, the [illuminated](https://www.abc-clio.com/ODLIS/odlis_i.aspx#illuminated) [manuscript](https://www.abc-clio.com/ODLIS/odlis_m.aspx#manuscript)s and [treasure binding](https://www.abc-clio.com/ODLIS/odlis_t.aspx#treasurebinding)s [exhibit](https://www.abc-clio.com/ODLIS/odlis_e.aspx#exhibit)ed by the Metropolitan Museum in New York City. Other items found in museums and of interest to [librarian](https://www.abc-clio.com/ODLIS/odlis_l.aspx#librarian)s include [inscription](https://www.abc-clio.com/ODLIS/odlis_i.aspx#inscription)s, [clay tablet](https://www.abc-clio.com/ODLIS/odlis_c.aspx#claytablet)s, [papyrus](https://www.abc-clio.com/ODLIS/odlis_p.aspx#papyrus) [scroll](https://www.abc-clio.com/ODLIS/odlis_s.aspx#scroll)s, [rare map](https://www.abc-clio.com/ODLIS/odlis_r.aspx#raremap)s, [letter](https://www.abc-clio.com/ODLIS/odlis_l.aspx#letter)s, [diaries](https://www.abc-clio.com/ODLIS/odlis_d.aspx#diary), etc. Many museums maintain a [library](https://www.abc-clio.com/ODLIS/odlis_m.aspx#museumlib) on the [premises](https://www.abc-clio.com/ODLIS/odlis_p.aspx#premises) containing books and other [reference](https://www.abc-clio.com/ODLIS/odlis_r.aspx#refbook) materials pertinent to their collections and activities. The concept of "museum" is often broadly interpreted to include archaeological and historical monuments, aquaria, arboreta, botanical gardens, nature centers, etc. Federal [grant](https://www.abc-clio.com/ODLIS/odlis_g.aspx#grant) support is provided to museums through the [Institute of Museum and Library Services](https://www.abc-clio.com/ODLIS/odlis_i.aspx#imls) (IMLS). Museums and exhibits are listed by region and subject category in _[Yahoo!](http://dir.yahoo.com/Society_and_Culture/Museums_and_Exhibits/)_ For leading art museums and galleries, try the [Art Museum Network](http://www.amn.org/).

A good deal of overlap might be inferred. First, both museums and archives can focus on unpublished, one-of-a-kind materials. In addition, both can be designed for expert users and historians. Finally, neither really requires (though both can welcome) the printed word.

#### Museum Studies

A retired page on the Smithsonian website ([http://](http://museumstudies.si.edu/index.html)[museumstudies.si.edu](http://museumstudies.si.edu/index.html)[/](http://museumstudies.si.edu/index.html)[index.html](http://museumstudies.si.edu/index.html)) stated the following about the theoretical approaches to museum work:

For the purposes of this Module, this is a great definition—worth keeping even if the website can only be viewed through the Internet Archive's WayBack Machine (here's a screenshot from 2017: [https://web.archive.org/web/20170321065703/http://museumstudies.si.edu/index.html](https://web.archive.org/web/20170321065703/http://museumstudies.si.edu/index.html)).

**Convergence in LAM Institutions, But Not Their Approach to IO**
-----------------------------------------------------------------

Libraries, archives, and museums together are often referred to as LAM institutions. When “galleries” (i.e., art galleries) are added to the mix, we get GLAM institutions.

*   All three traditions have their own vocabularies. This makes it easy for an archivist to talk to another archivist but more difficult for that archivist to talk to a museum curator…
*   The OAIS Reference Model ([http://public.ccsds.org/publications/archive/650x0m2.pdf](http://public.ccsds.org/publications/archive/650x0m2.pdf)) tries to overcome this issue in the field of digital preservation
*   Overlapping education opportunities, especially between librarianship and archival studies, can help alleviate vocabulary issues through “cross-training”

There’s a fair bit of talk about “Convergence”—in terms of concerns, approaches, and solutions in these institutions, despite the differences we have.

All of these institutions have different approaches, different content, different users, different structures, and different practices.
