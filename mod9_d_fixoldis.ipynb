**What about Virtual Libraries?**

**What is a Digital Archives?**
-------------------------------

The page about digital libraries demonstrated that a long-term preservation systems can be termed a "digital archive"—but what about when "archive**s**" is provided in the plural? 

DPLA was previously given as an example of a digital library. In some ways, it might also be considered a digital archives. Why? Consider the definition of [digital archives](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalarchives) in the ODLIS: 

> [Archival](https://products.abc-clio.com/ODLIS/odlis_a.aspx#archives) materials that have been [converted](https://products.abc-clio.com/ODLIS/odlis_d.aspx#dataconversion) to [machine-readable](https://products.abc-clio.com/ODLIS/odlis_m.aspx#machinereadable) [format](https://products.abc-clio.com/ODLIS/odlis_f.aspx#format), usually for the sake of [preservation](https://products.abc-clio.com/ODLIS/odlis_p.aspx#preservation) or to make them more [accessible](https://products.abc-clio.com/ODLIS/odlis_a.aspx#accessibility) to users. A prime example is _[American Memory](http://memory.loc.gov/ammem/index.html)_, a project undertaken by the [Library of Congress](https://products.abc-clio.com/ODLIS/odlis_l.aspx#lc) to make [digital collections](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalcollec) of [primary sources](https://products.abc-clio.com/ODLIS/odlis_p.aspx#primarysource) on the history and culture of the United States available via the [Internet](https://products.abc-clio.com/ODLIS/odlis_i.aspx#internet). Also refers to [information](https://products.abc-clio.com/ODLIS/odlis_i.aspx#information) originally created in electronic format, preserved for its [archival value](https://products.abc-clio.com/ODLIS/odlis_a.aspx#archivalvalue) (see [digital archive](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalarchive)).

From looking at the DPLA and reading about it, you probably realize that much of the content is digitized holdings from archival collections. In this way, DPLA is arguably a digital archives, as well as a virtual library, both of which are generally considered to be kinds of digital libraries. 

Remember from the discussion of metadata categories and the terminology used: often these specialized terms are considered differently by different groups. The important thing for now is to understand the various ideas presented—and to appreciate that different communities within the information professions might mean slightly different things.

**What Is a Digital Repository?**
---------------------------------

Digital repositories often (but not always) contain text-based digital items relating to a specific institution. Some information professionals consider "digital repository" to be the overarching term, not "digital library." This is fine. To continue to stick with the ODLIS, please consider the following definition for [digital repository](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalrepository): 

> Many [academic](https://products.abc-clio.com/ODLIS/odlis_a.aspx#academiclib) and [research libraries](https://products.abc-clio.com/ODLIS/odlis_r.aspx#researchlib) are actively engaged in building [digital](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digital) [collection](https://products.abc-clio.com/ODLIS/odlis_l.aspx#libcollec)s of [books](https://products.abc-clio.com/ODLIS/odlis_b.aspx#book), [papers](https://products.abc-clio.com/ODLIS/odlis_p.aspx#paper), [theses](https://products.abc-clio.com/ODLIS/odlis_t.aspx#thesis), [media](https://products.abc-clio.com/ODLIS/odlis_m.aspx#media), and other [works](https://products.abc-clio.com/ODLIS/odlis_w.aspx#work) of interest to the institution served, as a means of [preserving](https://products.abc-clio.com/ODLIS/odlis_p.aspx#preservation) and disseminating scholarly [information](https://products.abc-clio.com/ODLIS/odlis_i.aspx#information). Usually locally [authored](https://products.abc-clio.com/ODLIS/odlis_a.aspx#author) or produced, [content](https://products.abc-clio.com/ODLIS/odlis_c.aspx#content) can be either [born digital](https://products.abc-clio.com/ODLIS/odlis_b.aspx#borndigital) or [reformatted](https://products.abc-clio.com/ODLIS/odlis_r.aspx#reformat). [Access](https://products.abc-clio.com/ODLIS/odlis_a.aspx#access) is generally unrestricted, in compliance with the [Open Archives Initiative](https://products.abc-clio.com/ODLIS/odlis_o.aspx#oai) (OAI) protocol for [metadata](https://products.abc-clio.com/ODLIS/odlis_m.aspx#metadata) [harvesting](https://products.abc-clio.com/ODLIS/odlis_h.aspx#harvesting), which makes such archives [interoperable](https://products.abc-clio.com/ODLIS/odlis_i.aspx#interoperability) and cross-searchable. ...

Repositories, in this case, contain scholarly information, that is generally local in nature. These are often termed "institutional repositories"—for obvious reasons. All of the content is generated by individuals associated with a particular institution, and can be on any number of topics and from any discipline.

Other kinds of digital repositories exist, too. 

**Disciplinary repositories** contain contributions from researchers at a number of institutions, not just one. All of the contributions will be from the same discipline (e.g., astronomy). 

#### Examples of Repositories

Institutional repositories are one of the most common examples of digital repositories. You have already looked at [MOspace](https://mospace.umsystem.edu/xmlui/) as an example of a retrieval system, but did you realize that the content included in MOspace is 1) entirely digital and 2) entirely related to the University of Missouri? This means it is not only a digital library, but it is an institutional repository, and one that supports long-term digital preservation.

Take another look at MOspace as an example of an institutional repository: [https://mospace.umsystem.edu/xmlui/](https://mospace.umsystem.edu/xmlui/) 

Do you know why MOspace has its name? The "MO" at the beginning is the abbreviation for the state of Missouri, but the "space" at the end—what is that? In fact, the "space" is in homage to the software that is used: DSpace, an open source system originally created at MIT.

Wikipedia provides the following as an introduction to [DSpace](https://en.wikipedia.org/wiki/DSpace):

> **DSpace** is an [open source](https://en.wikipedia.org/wiki/Open_source_software "Open source software") repository software package typically used for creating [open access](https://en.wikipedia.org/wiki/Open_access "Open access") repositories for scholarly and/or published digital content. While DSpace shares some feature overlap with [content management systems](https://en.wikipedia.org/wiki/Content_management_system "Content management system") and [document management systems](https://en.wikipedia.org/wiki/Document_management "Document management"), the DSpace repository software serves a specific need as a digital archives system, focused on the long-term storage, access and preservation of digital content.

Other examples of digital repositories that are NOT institutional repositories, all vetted by [OpenDOAR](http://v2.sherpa.ac.uk/opendoar/), include:

*   **Disciplinary repository** of the Mexican Center for Innovation in Geothermal Energy, CeMIEGeo: [https://colecciondigital.cemiegeo.org/xmlui/](https://colecciondigital.cemiegeo.org/xmlui/) 
*   **Disciplinary repository** (geographically-focused) providing access to the research output of academics based in African countries, Cameroun's CamPuce [http://eprints.campuce.org/](http://eprints.campuce.org/) 
*   **Governmental repository** of the Superior Court of Justice (STJ) of Brazil: [http://bdjur.stj.jus.br/jspui/](http://bdjur.stj.jus.br/jspui/) 
*   **Aggregating repository,** AIJR Preprints,  a multidisciplinary preprint server that publishes early research work (without peer-reviewing) from all fields of research out of India: [https://preprints.aijr.org/index.php/ap/preprints](https://preprints.aijr.org/index.php/ap/preprints)

**What Is a Data Repository?**
------------------------------

Data repositories are digital repositories that house **datasets**. These are often spreadsheets, but not always. These datasets may or may not be retained indefinitely. An institution may house datasets in the institutional repository or may maintain a separate repository for the datasets. 

Consider the following definition of [data](https://products.abc-clio.com/ODLIS/odlis_d.aspx#data) from the ODLIS:

> The plural of the Latin word _datum_, meaning "what is given," often used as a singular collective noun. Facts, figures, or instructions presented in a form that can be comprehended, interpreted, and communicated by a human being or [processed](https://products.abc-clio.com/ODLIS/odlis_d.aspx#dataprocessing) by a computer. Compare with [information](https://products.abc-clio.com/ODLIS/odlis_i.aspx#information) and [knowledge](https://products.abc-clio.com/ODLIS/odlis_jk.aspx#knowledge). **_See also_**: [data bank](https://products.abc-clio.com/ODLIS/odlis_d.aspx#databank), [database](https://products.abc-clio.com/ODLIS/odlis_d.aspx#database), [data set](https://products.abc-clio.com/ODLIS/odlis_d.aspx#dataset), and [metadata](https://products.abc-clio.com/ODLIS/odlis_m.aspx#metadata).

Data in this context is generally _research data_ or content created in support of research projects—the kind of content that can be analyzed in the quest for new knowledge. Oceanographers create data when they take the temperature of ocean waters and observe weather conditions and currents. Information scientists create data when they observe users of an interface attempting to search for high-quality information. 

The Registry of Research Data Repositories ([https://www.re3data.org/](https://www.re3data.org/)) gateway lists the following types of content that can be considered data or necessary for the study and use/re-use of research data: [Archived data](https://www.re3data.org/search?query=&contentTypes%5b%5d=Archived%20data), [Audiovisual data](https://www.re3data.org/search?query=&contentTypes%5b%5d=Audiovisual%20data), [Configuration data](https://www.re3data.org/search?query=&contentTypes%5b%5d=Configuration%20data), [Databases](https://www.re3data.org/search?query=&contentTypes%5b%5d=Databases), [Images](https://www.re3data.org/search?query=&contentTypes%5b%5d=Images), [Networkbased data](https://www.re3data.org/search?query=&contentTypes%5b%5d=Networkbased%20data), [Plain text](https://www.re3data.org/search?query=&contentTypes%5b%5d=Plain%20text), [Raw data](https://www.re3data.org/search?query=&contentTypes%5b%5d=Raw%20data), [Scientific and statistical data formats](https://www.re3data.org/search?query=&contentTypes%5b%5d=Scientific%20and%20statistical%20data%20formats), [Software applications](https://www.re3data.org/search?query=&contentTypes%5b%5d=Software%20applications), [Source code](https://www.re3data.org/search?query=&contentTypes%5b%5d=Source%20code), [Standard office documents](https://www.re3data.org/search?query=&contentTypes%5b%5d=Standard%20office%20documents), [Structured graphics](https://www.re3data.org/search?query=&contentTypes%5b%5d=Structured%20graphics), [Structured text](https://www.re3data.org/search?query=&contentTypes%5b%5d=Structured%20text), [other](https://www.re3data.org/search?query=&contentTypes%5b%5d=other)

If you are interested in scholarly communication and the way that information professionals support research data management (RDM) be sure to click on the links above to see examples from member repositories.

In this context, then, a data repository is synonymous with the ODLIS term [data bank:](https://products.abc-clio.com/ODLIS/odlis_d.aspx#databank)

> Sometimes used [synonym](https://products.abc-clio.com/ODLIS/odlis_s.aspx#synonym)ously with _[database](https://products.abc-clio.com/ODLIS/odlis_d.aspx#database)_, the [term](https://products.abc-clio.com/ODLIS/odlis_t.aspx#term) applies more specifically to a [collection](https://products.abc-clio.com/ODLIS/odlis_c.aspx#collection) of nonbibliographic [data](https://products.abc-clio.com/ODLIS/odlis_d.aspx#data), usually numeric (_example_: _[Child Trends DataBank](http://www.childtrendsdatabank.org/)_). Large data banks containing [information](https://products.abc-clio.com/ODLIS/odlis_i.aspx#information) about individuals (social security numbers, credit history, health records, etc.) have become the [subject](https://products.abc-clio.com/ODLIS/odlis_s.aspx#subject) of controversy as the rapid development of high-speed [information technology](https://products.abc-clio.com/ODLIS/odlis_i.aspx#infotech) poses new threats to personal [privacy](https://products.abc-clio.com/ODLIS/odlis_p.aspx#privacy).

**Interlude: Digital Object Identifiers**

Books have unique identifiers known as ISBNs—which is a fantastic way to differentiate between books with similar titles, different editions, etc. Disambiguation is important, and unique identifiers are incredibly helpful in this regard.

Digital repository artifacts, especially articles and datasets, now are being assigned (i.e., being _minted_) DOIs. What is a DOI? According to the ODLIS, a [Digital Object Identifier (DOI)](https://products.abc-clio.com/ODLIS/odlis_d.aspx#doi) is the following:

> A unique code preferred by publishers in the identification and exchange of the content of a [digital object](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalobject), such as a [journal](https://products.abc-clio.com/ODLIS/odlis_jk.aspx#journal) [article](https://products.abc-clio.com/ODLIS/odlis_a.aspx#article), [Web](https://products.abc-clio.com/ODLIS/odlis_w.aspx#www) [document](https://products.abc-clio.com/ODLIS/odlis_d.aspx#document), or other item of [intellectual property](https://products.abc-clio.com/ODLIS/odlis_i.aspx#intellectualproperty). The DOI consists of two parts: a prefix assigned to each publisher by the administrative DOI agency and a suffix assigned by the publisher that may be any code the publisher chooses. DOIs and their corresponding [URL](https://products.abc-clio.com/ODLIS/odlis_u.aspx#url)s are registered in a central DOI directory that functions as a [routing](https://products.abc-clio.com/ODLIS/odlis_r.aspx#routing) system.
> 
> The DOI is [_persistent_](https://products.abc-clio.com/ODLIS/odlis_p.aspx#persistence), meaning that the identification of a digital object does not change even if ownership of or [rights](https://products.abc-clio.com/ODLIS/odlis_r.aspx#rights) in the entity are transferred. It is also _actionable_, meaning that clicking on it in a [Web browser](https://products.abc-clio.com/ODLIS/odlis_w.aspx#webbrowser) display will redirect the user to the content. The DOI is also [_interoperable_](https://products.abc-clio.com/ODLIS/odlis_i.aspx#interoperability), designed to function in past, present, and future [digital](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digital) technologies. The registration and resolver system for the DOI is run by the [International DOI Foundation](http://www.doi.org/) (IDF). [_CrossRef_](http://www.crossref.org/) is a collaborative [citation](https://products.abc-clio.com/ODLIS/odlis_c.aspx#citation) [linking](https://products.abc-clio.com/ODLIS/odlis_l.aspx#link) service that uses the DOI. ...

The purpose of a DOI is to provide a persistent identifier for an information resource. If the location of the resource changes, the DOI remains the same; what changes is the URL at DOI agency. When you click on a DOI you are sent to the DOI agency which then redirect your search to the digital repository in which the digital resource is stored.

When you provide a reference for an article in APA, you will generally include a DOI at the end of the reference when one is available.
