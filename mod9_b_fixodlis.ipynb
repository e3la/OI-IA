Key Ideas in IO: Digital Libraries as Systems
=============================================

The last module looked closely at online library catalogs as systems in which metadata is instrumental in providing access to library content (often, books) to library catalog users. This page and the next few pages focus on another kind of system used in the information professions: the digital library.

**What Is a Digital Library?**
------------------------------

Digital libraries are organized collections of digital materials. As with catalogs, information intermediaries, namely metadata librarians in this case, apply standards as a way of providing access to specific digital (and digitized) content.

As with other kinds of collections, the contents of digital libraries are usually vetted, and it is up to the information intermediaries to implement IO principles and practices accurately and consistently to make that content both retrievable and usable.

The ODLIS defines [digital libraries](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitallibrary) in the following way:

> A [library](https://products.abc-clio.com/ODLIS/odlis_l.aspx#library) in which a significant proportion of the resources are available in [machine-readable](https://products.abc-clio.com/ODLIS/odlis_m.aspx#machinereadable) format (as opposed to [print](https://products.abc-clio.com/ODLIS/odlis_p.aspx#printing) or [microform](https://products.abc-clio.com/ODLIS/odlis_m.aspx#microform)), accessible by means of computers. The [digital](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digital) content may be locally held or [accessed remotely](https://products.abc-clio.com/ODLIS/odlis_r.aspx#remoteaccess) via computer [networks](https://products.abc-clio.com/ODLIS/odlis_n.aspx#network). For a distributed example, see the [Mountain West Digital Library](http://mwdl.org/) established by the Utah Academic Library Consortium. In libraries, the process of [digitization](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitization) began with the [catalog](https://products.abc-clio.com/ODLIS/odlis_c.aspx#catalog), moved to [periodical indexes](https://products.abc-clio.com/ODLIS/odlis_p.aspx#periodicalindex) and [abstracting services](https://products.abc-clio.com/ODLIS/odlis_a.aspx#abstractingservice), then to periodicals and large reference works, and finally to book publishing. Abbreviated _d-lib_. Compare with [virtual library](https://products.abc-clio.com/ODLIS/odlis_v.aspx#virtuallibrary). **_See also_**: [Digital Library Federation](https://products.abc-clio.com/ODLIS/odlis_d.aspx#dlf), [Digital Public Library of America](https://products.abc-clio.com/ODLIS/odlis_d.aspx#dpla), and [National Science Digital Library](https://products.abc-clio.com/ODLIS/odlis_n.aspx#nsdl).

Unsurprisingly, digital library content is not printed on paper or viewable on microfiche. The content is stored in the digital library system and is retrievable, generally with just a click or two from the surrogate record.

Focusing on the sentence in the middle of the above example:

> In libraries, the process of [digitization](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitization) began with the [catalog](https://products.abc-clio.com/ODLIS/odlis_c.aspx#catalog), moved to [periodical indexes](https://products.abc-clio.com/ODLIS/odlis_p.aspx#periodicalindex) and [abstracting services](https://products.abc-clio.com/ODLIS/odlis_a.aspx#abstractingservice), then to [periodicals](https://products.abc-clio.com/ODLIS/odlis_p.aspx#periodical) and large [reference works](https://products.abc-clio.com/ODLIS/odlis_r.aspx#refbook), and finally to [book](https://products.abc-clio.com/ODLIS/odlis_b.aspx#book) [publishing](https://products.abc-clio.com/ODLIS/odlis_p.aspx#publishing).

Separating out the difference between digitized or digital content and digital libraries is worth the time.

Card catalog cards became digital when the content was entered into digital systems and encoded with MARC. An important aspect of making the print content digital was ensuring that it could be read by the computer with MARC encoding.

Such a statement is probably straightforward enough, but the sentence above is a little misleading. Those electronic bibliographic records are surrogates and are not the digitized resource. The catalog is not a digital library, then, according to the first sentence of the definition, because it does not include the digital content described by the surrogate records.

**Digital Library Software**
----------------------------

Online library catalogs use software that allows the storage and display of MARC records, and may include much more functionality if it is an integrated library system. Digital library software has some of the same functionality, but is in many cases different. If the digital library stores the digital resources, additional features are needed to support that function.

### **Examples of Digital Libraries**

##### **Digital Public Library of America—[https://dp.la/](https://dp.la/)**

Consider the Digital Public Library of America (DPLA). More about the initiative is available in Wikipedia: [https://en.wikipedia.org/wiki/Digital\_Public\_Library\_of\_America](https://en.wikipedia.org/wiki/Digital_Public_Library_of_America)

DPLA is a shining example of digital content available through a web-based interface. DPLA does not include the digital resources that it organizes. Rather it aggregates metadata from other digital library systems. It fits the definition of a digital library in that a "significant proportion of the resources are available in machine-readable format," even though those resources are stored elsewhere.

Information about participation in DPLA, including how to adjust your institution's metadata for inclusion, is available in the area of the website for information professionals: [https://pro.dp.la](https://pro.dp.la)

##### **Biodiversity Heritage Library—[https://www.biodiversitylibrary.org/](https://www.biodiversitylibrary.org/)**

The Biodiversity Heritage Library is another example of a digital library. It differs from DPLA in that it includes the digital content that its bibliographic records describe.

**What Is a Digital Archive (Digital Preservation System)?**
------------------------------------------------------------

The term "archive" can be used to mean a few different things. In the context of the digital content in the digital library it refers to the system that is used to ensure that the digital material will be preserved long-term. The ODLIS defines [digital archive](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalarchive) in the following way: 

> A system designed for locating, storing, and providing access to [digital](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digital) materials over the long term. A digital [archive](https://products.abc-clio.com/ODLIS/odlis_a.aspx#archive) may use a variety of [preservation](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalpreservation) methods to ensure that materials remain usable as technology changes, including [emulation](https://products.abc-clio.com/ODLIS/odlis_e.aspx#emulation) and [migration](https://products.abc-clio.com/ODLIS/odlis_m.aspx#migration). The [National Digital Information Infrastructure and Preservation Program](https://products.abc-clio.com/ODLIS/odlis_n.aspx#ndiipp) (NDIIPP) led by the [Library of Congress](https://products.abc-clio.com/ODLIS/odlis_l.aspx#lc) is an example of a program aimed at preserving digital [content](https://products.abc-clio.com/ODLIS/odlis_c.aspx#content). Compare with [digital archives](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalarchives). **_See also_**: [digital repository](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalrepository)

Libraries are not generally tasked with the long-term preservation of all the materials they house.  (Have you learned about "weeding" projects in libraries in your other classes yet?) Archives, unlike libraries, tend to be more interested in the long-term preservation of resources, and digital archives, accordingly, are sensitive to the need of long-term preservation.

What does this mean? For example, digital archives (and digital libraries, too) might be designed to conform to the requirements set forth by the OAIS (the [Open Archival Information System](https://en.wikipedia.org/wiki/Open_Archival_Information_System)) specifications initially put forth by NASA. The link above will take you to the Wikipedia page for the OAIS. If you have any interest at all in digital preservation, archives, or long-term access to digital or digitized content, you will likely want to take a look.

### **What Is the OAIS Reference Model and What Is an OAIS-Compliant System?**

OAIS (the [Open Archival Information System](https://en.wikipedia.org/wiki/Open_Archival_Information_System)) is a framework for thinking about what digital preservation systems need to be able to do. The ODLIS defines [OAIS](https://products.abc-clio.com/ODLIS/odlis_o.aspx#oais) in the following way: 

> A reference model for [digital archiving](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalarchive) systems initially developed by the Consultative Committee for Space Data Systems and adopted by [ISO](https://products.abc-clio.com/ODLIS/odlis_i.aspx#iso) as International [Standard](https://products.abc-clio.com/ODLIS/odlis_s.aspx#standard) ISO 14721:2003. The OAIS model is widely used by [libraries](https://products.abc-clio.com/ODLIS/odlis_l.aspx#library) as a framework for the development of [preservation](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalpreservation) [archives](https://products.abc-clio.com/ODLIS/odlis_a.aspx#archives) for [digital](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digital) materials. **_See also_**: [digital curation](https://products.abc-clio.com/ODLIS/odlis_d.aspx#digitalcuration).

Basically, content comes into the preservation system as a **Submission Information Package (SIP)**, is retained in the system as an **Archival Information Package (AIP)**, and is provided to users as a **Dissemination Information Package (DIP)**. This is explained a bit more [here](https://digitalarchive.mcmaster.ca/node/56), or in the OAIS standard itself.
