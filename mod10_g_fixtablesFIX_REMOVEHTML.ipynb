Metadata Elements and Relationships
===================================

Libraries have a great amount of authoritative data that has only recently begun to be restructured for linked data applications. You have seen that library data is structured data. It is not, however, structured for linked data use. Looking at some of the fields in the surrogate record, is it possible to re-envision them as linked data that is encoded in ways that computers can go beyond "reading" as in MARC (i.e., MAchine-_Readable_ Cataloging) to being **understood,** in all their complexity, by library systems and on the web of data _writ large_? 

Table 1 gives a snapshot of the surrogate record for the book, _Sea People_.

The record lists attributes of the book, such as author, title, publisher, and pagination. It serves a useful purpose in assisting users to Find, Identify, Select, and Obtain desired information resources. (Remember the first four [LRM User Tasks](https://www.isko.org/cyclo/lrm#3)?) In a library catalog, users can even click on some of the links and explore related material.

**What the surrogate record does not convey is relationships.** Relationships are a major way that users "explore" (the fifth and most elusive [LRM User Task](https://www.isko.org/cyclo/lrm#3)).

Current surrogate records are flat. They present descriptions of information resources by listing attributes. They do not convey the idea that the resources, when considered through the lens of relationships, are multi-dimensional, related to each other in complex and potentially unexpected ways.

What does it mean that the record does not convey relationships? This relationship map (Figure 1) visually shows some of the relationships held by the book and goes further and shows relationships that those entities have to other entities.

**Figure 1 A Sample Relationship Map for _Sea People_**

![Book-relationships.jpg](https://missouri.instructure.com/courses/49361/files/8633286/download)

Some of the relationships:

*   _Sea People_ was **written by a person**. That person **has a college degree** and, hence, a relationship to a school. She **has a teaching job** and family. She **took a trip** as she was writing the book.
*   _Sea People_ was **published by a company**. The company exists somewhere, publishes other material, and has employees, such as editors. Those entities have lots of relationships, too, none of which are included in the relationship map.
*   _Sea People,_ the printed book, is **related to the audio version**, which also has a relationship to the publisher.
*   _Sea People_ is **about Polynesians and Polynesia**, as well as other topics. It includes information about places, people, and events. It is related to other resources of the same genre and same topic.

Look at the surrogate record again. Our current online catalogs do well at listing information about our information resources. Our surrogate records, for example, give the name of the author that is listed on the book, but that has a different nuance than the richness that is Christina Thompson. Even with MARC encoding, asking surrogate records to convey the relationships inherent in the publication of a book or looking to them to help "explore" relationships is asking for more than they can deliver.

This data is not linked data. The library-created metadata in most of our library catalogs is not shared on the web and has not been extracted from other sources.

**Building a Web of Information**
---------------------------------

Once library data is converted to triples it can be used to build a web of information.

**But, Why Do This?**
---------------------

*   View Figure 1 again and envision a catalog that links to useful information in the catalog and from other sources.
*   Metadata elements encoded with MARC encoding cannot be used in non-library web applications. 
*   MARC-based online catalogs cannot make use of the more sophisticated technologies libraries could use to enhance user experiences.

*   If data elements in surrogate records were stored as triples and if the library catalog used linked data technology, library systems might re-use data stored in data stores. In addition, library users might be able to explore some of the relationships present in the record. For example, from the surrogate record for a book, the user could follow a link to more information about the author. From that record or site, the user could follow a link to information about the author's alma mater, the University of Melbourne. Conversely, the online catalog could display an information icon next to the author's name in the surrogate record and a pop-up box would provide information about the author from another source such as Wikidata. This is the beginning of making the surrogate record a player in a web of relevant and useful information.

The user experience will be very different when the information in a surrogate record has connections to a web of related information. For now, this module has focused on work in the cultural heritage sector. But, if you reflect on what the results semantic web/Web 3.0/web of data will will look when a critical mass of content has been created, the resulting network will be amazing!

<h2 class="mt-xl" style="color: #de5b49;"><strong>Building a Web of Information</strong></h2>
<p>Once library data is converted to triples it can be used to build a web of information.</p>
<h2 class="mt-xl" style="color: #de5b49;"><strong>But, Why Do This?</strong></h2>
<ul>
<li>View Figure 1 again and envision a catalog that links to useful information in the catalog and from other sources.</li>
<li>Metadata elements encoded with MARC encoding cannot be used in non-library web applications.&nbsp;</li>
<li>MARC-based online catalogs cannot make use of the more sophisticated technologies libraries could use to enhance user experiences.</li>
</ul>
<ul>
<li>If data elements in surrogate records were stored as triples and if the library catalog used linked data technology, library systems might re-use data stored in data stores. In addition, library users might be able to explore some of the relationships present in the record. For example, from the surrogate record for a book, the user could follow a link to more information about the author. From that record or site, the user could follow a link to information about the author's alma mater, the University of Melbourne. Conversely, the online catalog could display an information icon next to the author's name in the surrogate record and a pop-up box would provide information about the author from another source such as Wikidata. This is the beginning of making the surrogate record a player in a web of relevant and useful information.</li>
</ul>
<p>The user experience will be very different when the information in a surrogate record has connections to a web of related information. For now, this module has focused on work in the cultural heritage sector. But, if you reflect on what the results semantic web/Web 3.0/web of data will will look when a critical mass of content has been created, the resulting network will be amazing!</p>
